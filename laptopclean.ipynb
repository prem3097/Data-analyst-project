```python
#Importing the  required libraries.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Uploading the csv file as df.
df = pd.read_csv('/content/laptopData.csv')
```

```python
#Understanding the dataset using the info.
df.info()
```

```python
#Checking the null values:
df['OpSys'] = df['OpSys'].replace('No OS', np.nan)
df['Weight'] = df['Weight'].replace('?', np.nan)
df.isnull().sum()
```

```python
#Fill the null values using fillna.
x = df["Company"].mode()[0]
df.fillna({"Company": x}, inplace=True)
x = df["TypeName"].mode()[0]
df.fillna({"TypeName": x}, inplace=True)
x = df["Inches"].mode()[0]
df.fillna({"Inches": x}, inplace=True)
x = df["ScreenResolution"].mode()[0]
df.fillna({"ScreenResolution": x}, inplace=True)
x = df["Cpu"].mode()[0]
df.fillna({"Cpu": x}, inplace=True)
x = df["Ram"].mode()[0]
df.fillna({"Ram": x}, inplace=True)
x = df["Memory"].mode()[0]
df.fillna({"Memory": x}, inplace=True)
x = df["Gpu"].mode()[0]
df.fillna({"Gpu": x}, inplace=True)
x = df["OpSys"].mode()[0]
df.fillna({"OpSys": x}, inplace=True)
x = df["Weight"].mode()[0]
df.fillna({"Weight": x}, inplace=True)
x = df["Price"].mean()
df.fillna({"Price": x}, inplace=True)
```

```python
#Checking the null values.
df.isnull().sum()
```

```python
#Droping unwanted column.
df.drop("Unnamed: 0", axis=1, inplace=True)
```

```python
#Checking the duplicate values.
df.duplicated().sum()
```

```python
#Removing the duplicate values.
df.drop_duplicates(inplace=True)
```

```python
#Reseting the index value.
df.reset_index(drop=True, inplace=True)
```

```python
#Downloading the cleaned file.
df.to_excel("cleaned_data.xlsx")
df
```

```python
# 7. Correlation between screen size and price
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Inches', y='Price', data=df)
plt.title('Screen Size vs. Price')
plt.show()
```

```python
# 8. RAM vs. Price
plt.figure(figsize=(10, 6))
sns.lineplot(x='Ram', y='Price', data=df)
plt.title('RAM vs. Price')
plt.show()
```

```python
# 9. SSD vs. HDD effect on price
df['Storage_Type'] = df['Memory'].apply(lambda x: 'SSD' if 'SSD' in x else 'HDD')
plt.figure(figsize=(10, 6))
sns.barplot(x='Storage_Type', y='Price', data=df)
plt.title('SSD vs. HDD - Price Comparison')
plt.show()
```

```python
# 10. Gaming GPU effect on price
gaming_gpus = ['Nvidia', 'AMD']  # Adjust based on what's in your data
df['Is_Gaming_GPU'] = df['Gpu'].apply(lambda x: any(g in x for g in gaming_gpus))
plt.figure(figsize=(10, 6))
sns.barplot(x='Is_Gaming_GPU', y='Price', data=df)
plt.title('Gaming GPU vs. Price')
plt.xticks([0, 1], ['Non-Gaming', 'Gaming'])
plt.show()
```
